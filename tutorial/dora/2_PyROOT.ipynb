```python
# Example: displaying a ROOT histogram from Python
from ROOT import gRandom, TCanvas, TH1F
```

```python
c1 = TCanvas('c1', 'Example', 200, 10, 700, 500)
```

```python
hpx = TH1F('hpx', 'px', 100, -4 , 4)
```

```python
for i in xrange(25000):
    px = gRandom.Gaus()
    i = hpx.Fill(px)
```

```python
hpx.Draw()
```

```python
c1.Draw()
```
